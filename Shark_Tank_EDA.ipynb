### Q5.	How does the deal_equity distribution look? Are there any outliers?

```python
deal_done= df[df["deal"]==1]

plt.figure(figsize=(10, 5))
sns.boxplot(x=deal_done['deal_equity'], color='blue')
plt.title("Distribution of Deal Equity (%)")
plt.xlabel("Equity Percentage Offered in Deals")
plt.grid(True)
plt.show()

# IQR method to find outliers
Q1 = deal_done['deal_equity'].quantile(0.25)
Q3 = deal_done['deal_equity'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = deal_done[
    (deal_done['deal_equity'] < lower_bound) |
    (deal_done['deal_equity'] > upper_bound)
]

# Show outlier brands and their deal equity
print(outliers[['brand_name', 'deal_equity']])
```

### Q14.	Which shark has the highest return on investment (ROI) based on deal_amount vs. deal_equity?

```python
deals_df = df[df['deal'].astype(str).str.strip() == '1'].copy()
deals_df['amount_per_shark'] = deals_df['deal_amount'] / deals_df['total_sharks_invested']

deals_df['roi'] = deals_df['deal_equity'] / deals_df['amount_per_shark']
shark_columns = [
    'ashneer_deal', 'anupam_deal', 'aman_deal',
    'namita_deal', 'vineeta_deal', 'peyush_deal', 'ghazal_deal'
]

shark_rois = {}
for shark in shark_columns:
    shark_deals = deals_df[deals_df[shark] == 1]
    avg_roi = shark_deals['roi'].mean()
    # Format shark name (capitalize first letter)
    shark_name = shark.replace('_deal', '').capitalize()
    shark_rois[shark_name] = avg_roi

sorted_shark_rois = dict(sorted(shark_rois.items(), key=lambda x: x[1], reverse=True))

print("Average ROI per Shark (Equity % per â‚¹ Lakh Invested):")
for shark, roi in sorted_shark_rois.items():
    print(f"{shark}: {roi:.2f}")
```

### Q17.	Identify if there is a pattern in episode-wise deal closuresâ€”do certain episodes see more investments?

### Q18.	Analyze the impact of equity dilution on deal_valuation. Are lower equity asks leading to higher deal valuations?

```python
deals_df['rounded_equity'] = deals_df['deal_equity'].round()
avg_valuation_by_equity = deals_df.groupby('rounded_equity')['deal_valuation'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.lineplot(data=avg_valuation_by_equity, x='rounded_equity', y='deal_valuation', marker='o', color='green')
plt.title('Average Deal Valuation vs Equity Given')
plt.xlabel('Equity Given (%)')
plt.ylabel('Average Deal Valuation (â‚¹ lakh)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
